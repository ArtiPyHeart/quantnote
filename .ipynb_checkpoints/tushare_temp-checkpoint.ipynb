```python
import tushare as ts
import pandas as pd
import numpy as np
from tqdm import tqdm
import arrow
from glob import glob
import time
import os
from tenacity import retry, stop_after_attempt, wait_fixed
```

```python
ts_api = ts.pro_api('185f21b39d2eb92b35c6b41a7e4bdb90998faedd249042147af2b24f')
data_path = '/Volumes/WD/tushare/'
```

```python
ts_api.daily(trade_date='20200417')
```

```python
trade_date = ts_api.trade_cal(exchange='SSE', is_open='1', start_date='20190101', end_date='20191231', fields='cal_date')
trade_date.head()
```

```python
def my_before_sleep(retry_state):
    print(f'Retrying {retry_state.fn}: attempt {retry_state.attempt_number} ended with: {retry_state.outcome}')

@retry(wait=wait_fixed(1), stop=stop_after_attempt(50), before_sleep=my_before_sleep)
def get_data(date):
    df = ts_api.daily(trade_date=date)
    return df
```

```python
for d in tqdm(reversed(trade_date['cal_date'].tolist())):
    time.sleep(0.5)
    full_path = data_path + f"stock_{d}.pkl"
    if os.path.exists(full_path):
        continue
    else:
        get_data(d).to_pickle(full_path)
```

```python
data_ls = sorted(glob(data_path + 'stock_*.pkl'))
print(data_ls[:1])
pd.read_pickle(data_ls[0]).head()
```

```python
index = pd.read_excel('000985中证全指.xlsx', dtype={'成分券代码Constituent Code': str})
index.head()
```

```python
code_to_name = index.set_index('成分券代码Constituent Code')['成分券名称Constituent Name'].to_dict()
code_to_name['000001']
```

```python
tmp = pd.read_pickle(data_ls[0])
tmp['ts_code'] = tmp['ts_code'].str[:6]
tmp = tmp[tmp['ts_code'].isin(index['成分券代码Constituent Code'])]
tmp
```

```python
arrow.get(tmp['trade_date'].unique()[0]).date()
```

```python
stock_universe = set(pd.read_pickle(data_ls[0])['ts_code'].str[:6].tolist()) & set(index['成分券代码Constituent Code'].tolist())

stock_dict = {key: list() for key in stock_universe}
time_index = list()
for path in tqdm(data_ls):
    df = pd.read_pickle(path)
    df['ts_code'] = df['ts_code'].str[:6]
    time_index.append(arrow.get(df['trade_date'].unique()[0]).date())
    close_dict = df.set_index('ts_code')['close'].to_dict()
    for key in stock_universe:
        stock_dict[key].append(close_dict.get(key))

final_df = pd.DataFrame(data=stock_dict, index=time_index)
final_df.head()
```

```python
final_df = final_df.sort_index(axis=1).rename(columns=code_to_name)
```

```python
final_df.fillna(method='ffill').pct_change().dropna(how='all').to_pickle('a_return_pct.pkl')
```
